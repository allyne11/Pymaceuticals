```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame

combined_data = pd.merge(study_results, mouse_metadata, on="Mouse ID")
# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
num_mice = combined_data["Mouse ID"].nunique()
print(f"Total number of unique mice: {num_mice}")
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mice = combined_data[combined_data.duplicated(subset=["Mouse ID", "Timepoint"], keep=False)]

duplicate_mice
```

```python
# Identify the Mouse IDs that have duplicate Timepoints
duplicate_mouse_ids = duplicate_mice["Mouse ID"].unique()
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_data[~combined_data["Mouse ID"].isin(duplicate_mouse_ids)]

clean_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
num_mice2 = clean_data["Mouse ID"].nunique()
print(f"Total number of unique mice: {num_mice2}")
```

## Summary Statistics

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
summary_stats = clean_data.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(
    Mean="mean",
    Median="median",
    Variance="var",
    Standard_Deviation="std",
    SEM="sem"
).reset_index()

summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Select the four treatment regimens of interest
selected_regimens = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Filter the data for the selected regimens
filtered_data = clean_data[clean_data["Drug Regimen"].isin(selected_regimens)]

# Get the last (greatest) timepoint for each mouse
max_timepoints = filtered_data.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge this with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume = pd.merge(max_timepoints, filtered_data, on=["Mouse ID", "Timepoint"], how="left")

# Display the resulting DataFrame
final_tumor_volume.head()
```

```python
# Put treatments into a list for looping and plotting labels
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create an empty dictionary to store tumor volume data for each treatment
tumor_volumes = {}

# Calculate the IQR and determine potential outliers for each treatment regimen
outliers_dict = {}

for treatment in treatments:
    # Get tumor volumes for the current treatment
    treatment_data = final_tumor_volume.loc[final_tumor_volume["Drug Regimen"] == treatment, "Tumor Volume (mm3)"]
    
    # Store tumor volume data for later plotting
    tumor_volumes[treatment] = treatment_data
    
    # Calculate quartiles and IQR using scipy.stats
    lowerq, median, upperq = st.mstats.mquantiles(treatment_data, prob=[0.25, 0.5, 0.75])
    iqr = upperq - lowerq

    # Determine bounds for outliers
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)

    # Identify outliers
    outliers = treatment_data[(treatment_data < lower_bound) | (treatment_data > upper_bound)]
    outliers_dict[treatment] = outliers

# Convert outliers dictionary to a DataFrame for display
outliers_df = pd.DataFrame.from_dict(outliers_dict, orient="index").transpose()

# Display the DataFrame
print("Potential Outliers in Tumor Volumes:")
print(outliers_df)
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
# Create a list of tumor volume data for each treatment group
tumor_data = [tumor_volumes[treatment] for treatment in treatments]

# Generate the box plot
plt.figure(figsize=(10, 6))
plt.boxplot(tumor_data, labels=treatments, patch_artist=True)

# Label the plot
plt.xlabel("Drug Regimen")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Distribution of Tumor Volume for Each Treatment Group")

# Show the plot
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
# Select a single mouse treated with Capomulin
capomulin_mice = clean_data[clean_data["Drug Regimen"] == "Capomulin"]
single_mouse_id = capomulin_mice["Mouse ID"].iloc[0]  # Select the first mouse in the dataset

# Filter data for the selected mouse
single_mouse_data = capomulin_mice[capomulin_mice["Mouse ID"] == single_mouse_id]

# Generate the line plot
plt.figure(figsize=(8, 6))
plt.plot(single_mouse_data["Timepoint"], single_mouse_data["Tumor Volume (mm3)"], marker='o', linestyle='-', color='b')

# Label the plot
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title(f"Tumor Volume Over Time for Mouse {single_mouse_id} (Capomulin Treatment)")

# Show the plot
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
# Filter data for Capomulin regimen
capomulin_data = clean_data[clean_data["Drug Regimen"] == "Capomulin"]

# Calculate the average tumor volume for each mouse
avg_tumor_volume = capomulin_data.groupby("Mouse ID")["Tumor Volume (mm3)"].mean().reset_index()

# Merge with mouse weight data
mouse_weights = capomulin_data[["Mouse ID", "Weight (g)"]].drop_duplicates()
capomulin_avg = pd.merge(avg_tumor_volume, mouse_weights, on="Mouse ID")

# Generate the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(capomulin_avg["Weight (g)"], capomulin_avg["Tumor Volume (mm3)"], color='b', alpha=0.7)

# Label the plot
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)")

# Show the plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
# Import necessary library for correlation and regression
from scipy.stats import linregress

# Calculate the correlation coefficient
correlation = capomulin_avg["Weight (g)"].corr(capomulin_avg["Tumor Volume (mm3)"])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(capomulin_avg["Weight (g)"], capomulin_avg["Tumor Volume (mm3)"])

# Generate regression line values
regression_line = capomulin_avg["Weight (g)"] * slope + intercept

# Generate the scatter plot with regression line
plt.figure(figsize=(8, 6))
plt.scatter(capomulin_avg["Weight (g)"], capomulin_avg["Tumor Volume (mm3)"], color='b', alpha=0.7, label="Data Points")
plt.plot(capomulin_avg["Weight (g)"], regression_line, color='r', label=f"y = {slope:.2f}x + {intercept:.2f}")

# Label the plot
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title(f"Mouse Weight vs. Average Tumor Volume (Capomulin Regimen)\nCorrelation: {correlation:.2f}")
plt.legend()

# Show the plot
plt.show()

# Display correlation coefficient and regression statistics
correlation, slope, intercept, r_value**2, p_value, std_err
```
